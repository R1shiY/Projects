```python
def check_record(weight_class, squat, bench, deadlift, records = records):
    # Filter records by weight class
    filtered = records[records["Class"] == weight_class]

    # Initialize results for each lift
    results = {"Squat": squat, "Bench": bench, "Deadlift": deadlift}
    updates = []
    off_by = {}

    # Iterate over each lift type
    for event, user_lift in results.items():
        # Get current records for the event
        event_records = filtered[filtered["Event"] == f"{event} (Full Power)"].sort_values(by="Lift", ascending=False)

        # If no records exist for the event in this class
        if event_records.empty:
            updates.append({
                "Event": f"{event} (Full Power)",
                "Class": weight_class,
                "Rank": 1,
                "Lifter": "You",
                "Lift": user_lift,
                "Date": "2024-11-25"  # Assuming today's date
            })
            continue

        # Extract record lifts
        ranks = event_records["Lift"].values

        # Compare user's lift to the records
        if user_lift > ranks[0]:
            updates.append({
                "Event": f"{event} (Full Power)",
                "Class": weight_class,
                "Rank": 1,
                "Lifter": "You",
                "Lift": user_lift,
                "Date": "2024-11-25"
            })
        elif len(ranks) > 1 and user_lift > ranks[1]:
            updates.append({
                "Event": f"{event} (Full Power)",
                "Class": weight_class,
                "Rank": 2,
                "Lifter": "You",
                "Lift": user_lift,
                "Date": "2024-11-25"
            })
        elif len(ranks) > 2 and user_lift > ranks[2]:
            updates.append({
                "Event": f"{event} (Full Power)",
                "Class": weight_class,
                "Rank": 3,
                "Lifter": "You",
                "Lift": user_lift,
                "Date": "2024-11-25"
            })
        else:
            closest = ranks[2] if len(ranks) > 2 else ranks[-1]
            off_by[event] = abs(user_lift - closest)

    # Update records DataFrame if needed
    if updates:
        new_records = pd.DataFrame(updates)
        records = pd.concat([records, new_records], ignore_index=True)
        return f"Congratulations! You set a record in: {', '.join([u['Event'] for u in updates])}", records
    else:
        closest_event = min(off_by, key=off_by.get)
        return f"You did not set a record. You were {off_by[closest_event]:.1f} off in {closest_event}.", records
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'data' is your provided dataset
# Ensure the dataset has columns: Event, Class, Rank, Lift
data = pd.DataFrame(records)
# Preprocess data to pivot for easier analysis
def preprocess_data(data):
    pivot_data = data.pivot_table(
        index=["Class"],
        columns=["Event", "Rank"],
        values="Lift",
        aggfunc="max"
    )
    pivot_data.columns = [f"{col[0]} Rank {col[1]}" for col in pivot_data.columns]
    pivot_data.reset_index(inplace=True)
    return pivot_data

# 1. Analyze Strength Trends
def analyze_strength_trends(data):
    print("Average Performance by Weight Class:")
    avg_performance = data.groupby("Class")["Lift"].mean()
    print(avg_performance)
    avg_performance.plot(kind="bar", figsize=(12, 6))
    plt.title("Average Performance by Weight Class")
    plt.ylabel("Lift Weight (lbs)")
    plt.xlabel("Weight Class")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 2. Find Correlations
def find_correlations(pivot_data):
    print("Correlation Matrix:")
    corr = pivot_data.corr()
    print(corr)
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Matrix")
    plt.show()
```

```python
# 3. Predict Performance
def predict_performance(data):
    features = data[["Class", "Lift"]]
    target = data["Lift"]

    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    print("Model Performance:")
    print(f"Mean Squared Error: {mean_squared_error(y_test, predictions):.2f}")
    print(f"R-squared: {r2_score(y_test, predictions):.2f}")

    return model
```

```python
# 4. Visualize Records and Outliers
def visualize_records(data):
    plt.figure(figsize=(12, 6))
    sns.boxplot(x="Class", y="Lift", data=data)
    plt.title("Lift Distribution by Weight Class")
    plt.ylabel("Lift (lbs)")
    plt.xlabel("Weight Class")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 5. Simulate a Virtual Competition
def simulate_virtual_competition(model, weight_class, lift, data):
    simulated_total = model.predict([[weight_class, lift]])[0]
    print(f"Simulated Lift Total for {weight_class} Class: {simulated_total:.2f}")

    # Create a new entry for the simulated lifter
    simulated_entry = pd.DataFrame({"Class": [weight_class], "Lift": [simulated_total]})

    # Append the simulated lifter to the dataset using pd.concat
    simulated_data = pd.concat([data, simulated_entry], ignore_index=True)

    # Sort the dataset by Lift in descending order and reset index
    simulated_data = simulated_data.sort_values(by="Lift", ascending=False).reset_index(drop=True)

    # Determine the rank of the new entry
    rank = simulated_data[simulated_data["Lift"] == simulated_total].index[0] + 1
    print(f"Your simulated rank in the {weight_class} class: {rank}")

    return simulated_data
```

```python
# Run the analysis
print("\nRunning Powerlifting Performance Analysis...\n")

# Preprocess data for pivoted structure
pivot_data = preprocess_data(data)

# Analyze trends
analyze_strength_trends(data)

# Find correlations
find_correlations(pivot_data)

# Train a model for predictions
model = predict_performance(data)

# Visualize records and outliers
visualize_records(data)

# Simulate a virtual competition
updated_data = simulate_virtual_competition(
    model,
    weight_class=123,
    lift=728,  # Example lift
    data=data
)
```
